```python
unpivot_df = df.melt(id_vars=['Area', 'Item', 'Element'], var_name='Year', value_name='Value')
```

```python
df.head()
```

```python
filtered_df = unpivot_df[unpivot_df['Element'].isin(['Export Quantity', 'Import Quantity', 'Production'])]

# Convert the 'Value' column to numeric (in case it's not already)
filtered_df['Value'] = pd.to_numeric(filtered_df['Value'], errors='coerce')

# Pivot the DataFrame to have elements as columns and years as rows
pivot_df = filtered_df.pivot_table(index=['Area', 'Item', 'Unit', 'Year'], columns='Element', values='Value').reset_index()

# Calculate the SSR using the formula
pivot_df['SSR'] = (pivot_df['Production'] * 100) / (pivot_df['Production'] + pivot_df['Import Quantity'] - pivot_df['Export Quantity'])

# Now you can analyze the trends over time
# For example, to see the SSR trend over the years for a specific item:
item_name = 'Almonds, in shell'
item_df = pivot_df[pivot_df['Item'] == item_name]

# Plot SSR trend over the years for the item
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(item_df['Year'], item_df['SSR'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Self Sufficiency Level (SSR)')
plt.title(f'SSR Trend for {item_name}')
plt.grid(True)
plt.show()
```
